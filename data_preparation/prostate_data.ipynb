```python
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np
import torch

def convert_amos_label(input_label):
    input_label_5 = (input_label == 5).float()
    input_label_4 = (input_label == 4).float()
    return input_label_5 + input_label_4

img = nib.load('/home/xiangcen/xiaoyao/data_prostate/003137_mask.nii')
img = torch.tensor(img.get_fdata())
print(img.max(), img.min(), img.shape)

print(type(img))
print(np.unique(img))
plt.imshow(convert_amos_label(img[:, :, 10]))
plt.colorbar()


median = 43//2
low = median - 18
high = median + 18


low, high
```
